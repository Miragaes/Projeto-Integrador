Abaixo estão as principais funções que serão utilizadas para formatar o arquivo:

def imprime_espaco -> função para imprimir espaços ' ' nos campos que estão vazios ou para completar o restante do campo, caso a quantidade de caracteres máxima do campo de texto ainda não tenha sido atingida.

def imprime_espaco -> função para imprimir zeros '0' nos campos que estão numericos caso a quantidade de caracteres máxima do campo de numerico ainda não tenha sido atingida.

def tratamento_ponto_campo_numerico -> como o padrão de leitura em cobol, o campo númerico que contém ponto (caso seu número esteja no padrão americano) é passado apenas com números, sem a utilizado de ponto, essa função irá fazer a retirada desse caractere para poder formatar o campo.

def tratamento_virgula_campo_numerico -> como o padrão de leitura em cobol, o campo númerico que contém vírugla (caso seu número esteja no padrão americano) é passado apenas com números, sem a utilizado da vírgula, essa função irá fazer a retirada desse caractere para poder formatar o campo.

def converte_campo_numerico_simples -> caso sua célula na planilha seja um número que não utiliza de '.' ou ',', está função irá converter o seu número para o padrão de entrada do arquivo em cobol.

def converte_campo_numerico_flutuante -> caso sua célula na planilha seja um número que utiliza de '.' ou ',', está função irá converter o seu número para o padrão de entrada do arquivo em cobol.

def converte_campo_alfanumerico -> caso sua célula na planilha seja um elemento alfanumerico, está função irá converter o seu campo para o padrão de entrada do arquivo em cobol.

def verifica_campo_vazio -> caso sua célula não contenha nenhum valor informado e esteja vazia/ não preenchida, está função irá tratar esse acontecimento e substituir o campo por ' ', repetindo a quantida de ' ' de acordo com o tamanho máximo que o campo determinado pelo arquivo vsam.

def converte-data -> faz a retirada de '/' e '-' caso a data contenha, e ':' em caso de horas, e deixa na formatação de apenas numeral.

```python
def imprime_espaco(quantidade):
    espacos = ''
    contador = 0
    while(contador < quantidade):
      espacos = espacos + " "
      contador = contador + 1
    
    return espacos

def imprime_zero(quantidade):
    zeros = ''
    contador = 0
    while(contador < quantidade):
      zeros = zeros + "0"
      contador = contador + 1
    
    return zeros

def tratamento_ponto_campo_numerico(valor):
    return valor.replace('.','')

def tratamento_virgula_campo_numerico(valor):
    return valor.replace(',','')

def tratamento_simbolo_monetario(valor):
    return valor.replace("R$",'')

def converte_campo_numerico_simples(valor, tamanho_campo):
    campo = ""
    if(len(valor) < tamanho_campo):
      campo = campo + imprime_zero(tamanho_campo - len(valor))
    campo = campo + valor

    return campo

def converte_campo_numerico_flutuante(valor, tamanho_campo):
    campo = ""

    if('.' in valor):
      valor = tratamento_ponto_campo_numerico(valor)
    if(',' in valor):
      valor = tratamento_virgula_campo_numerico(valor)
    if("R$" in valor):
      valor = tratamento_simbolo_monetario(valor)
    if(" " in valor):
      valor = valor.replace(" ",'')

    if(len(valor) < tamanho_campo):
      campo = campo + imprime_zero(tamanho_campo - len(valor))
    campo = campo + valor

    return campo

def converte_campo_alfanumerico(valor, tamanho_campo):
    campo = ""
    campo = campo + valor
    if(len(valor) < tamanho_campo):
      campo = campo + imprime_espaco(tamanho_campo - len(valor))

    return campo

def converte_data(valor, tamanho_campo):
  campo = ""
  if('/' in valor):
    valor = valor.replace('/','')
  if('-' in valor):
    valor = valor.replace('-','')
  if(':' in valor):
    valor = valor.replace(':','')
  if(len(valor) < tamanho_campo):
      campo = campo + imprime_zero(tamanho_campo - len(valor))
  
  return campo


def verifica_campo_vazio(valor, tamanho_campo):
    campo = ""
    if(len(valor) == 0):
      campo = campo + imprime_espaco(tamanho_campo)
    
    return campo
```

_ É utilizado a função str para trabalhar sempre com textos, indepente do campo que seja, para melhor manipulação dos dados
_ Como citado anterior, caso sua célula na tabela esteja vazia, é atribui espaço ' ' para ela no arquivo, caso não queria que isso seja feito, e mantenha o campo sem nada no arquivo de entra, basta comentar a linha que contém o comando 'else' e sua linha subsequente utilizando # no começo da linha

```python
contador = 0
while contador < tabela['Data Emissao Pedido'].size:
  linha = "w"
  
  if(verifica_campo_vazio(str(tabela['Data Emissao Pedido'][contador]), data_emissao_size ) == ""):
    linha = linha + converte_campo_numerico_simples(str(tabela['Data Emissao Pedido'][contador]), data_emissao_size)
  else:
    linha = linha + verifica_campo_vazio(str(tabela['Data Emissao Pedido'][contador]), data_emissao_size )

  if(verifica_campo_vazio(str(tabela['Status Pedido'][contador]), status_size) == ""):
    linha = linha + converte_campo_numerico_simples(str(tabela['Status Pedido'][contador]), status_size)
  else:
    linha = linha + verifica_campo_vazio(str(tabela['Status Pedido'][contador]), status_size)

  if(verifica_campo_vazio(str(tabela['Vendedor'][contador]), vendedor_size) == ""):
    linha = linha + converte_campo_numerico_simples(str(tabela['Vendedor'][contador]), vendedor_size)
  else:
    linha = linha + verifica_campo_vazio(str(tabela['Vendedor'][contador]), vendedor_size)

  if(verifica_campo_vazio(str(tabela['Canal do Pedido '][contador]), canal_size) == ""):
    linha = linha + converte_campo_numerico_simples(str(tabela['Canal do Pedido '][contador]), canal_size)
  else:
    linha = linha + verifica_campo_vazio(str(tabela['Canal do Pedido'][contador]), canal_size)

  if(verifica_campo_vazio(str(tabela['Id Cliente'][contador]), id_cliente_size) == ""):
    linha = linha + converte_campo_numerico_simples(str(tabela['Id Cliente'][contador]), id_cliente_size)
  else:
    linha = linha + verifica_campo_vazio(str(tabela['Id Cliente'][contador]), id_cliente_size)

  if(verifica_campo_vazio(str(tabela['Forma de Entrega'][contador]), entrega_size) == ""):
    linha = linha + converte_campo_numerico_simples(str(tabela['Forma de Entrega'][contador]), entrega_size)
  else:
    linha = linha + verifica_campo_vazio(str(tabela['Forma de Entrega'][contador]), entrega_size)

  if(verifica_campo_vazio(str(tabela['Data de Entrega'][contador]), data_entrega_size) == ""):
    linha = linha + converte_campo_numerico_simples(str(tabela['Data de Entrega'][contador]), data_entrega_size)
  else:
    linha = linha + verifica_campo_vazio(str(tabela['Data de Entrega'][contador]), data_entrega_size)

  if(verifica_campo_vazio(str(tabela['Valor da Entrega Agendada (Frete)'][contador]), frete_size) == ""):
    linha = linha + converte_campo_numerico_flutuante(str(tabela['Valor da Entrega Agendada (Frete)'][contador]), frete_size)
  else:
    linha = linha + verifica_campo_vazio(str(tabela['Valor da Entrega Agendada (Frete)'][contador]), frete_size)

  if(verifica_campo_vazio(str(tabela['Id Produto'][contador]), id_produto_size) == ""):
    linha = linha + converte_campo_numerico_simples(str(tabela['Id Produto'][contador]), id_produto_size)
  else:
    linha = linha + verifica_campo_vazio(str(tabela['Id Produto'][contador]), id_produto_size)

  if(verifica_campo_vazio(str(tabela['Valor Unitario Produto'][contador]), valor_unitario_size) == ""):
    linha = linha + converte_campo_numerico_flutuante(str(tabela['Valor Unitario Produto'][contador]), valor_unitario_size)
  else:
    linha = linha + verifica_campo_vazio(str(tabela['Valor Unitario Produto'][contador]), valor_unitario_size)

  if(verifica_campo_vazio(str(tabela['Qtd Produto'][contador]), qtd_produto_size) == ""):
    linha = linha + converte_campo_numerico_simples(str(tabela['Qtd Produto'][contador]), qtd_produto_size)
  else:
    linha = linha + verifica_campo_vazio(str(tabela['Qtd Produto'][contador]), qtd_produto_size)

  if(verifica_campo_vazio(str(tabela['Total Produto'][contador]), total_produto_size) == ""):
    linha = linha + converte_campo_numerico_flutuante(str(tabela['Total Produto'][contador]), total_produto_size)
  else:
    linha = linha + verifica_campo_vazio(str(tabela['Total Produto'][contador]), total_produto_size)

  if(verifica_campo_vazio(str(tabela['Intensao de Forma de Pagamento'][contador]), forma_pagamento_size) == ""):
    linha = linha + converte_campo_numerico_simples(str(tabela['Intensao de Forma de Pagamento'][contador]), forma_pagamento_size)
  else:
    linha = linha + verifica_campo_vazio(str(tabela['Intensao de Forma de Pagamento'][contador]), forma_pagamento_size)

  if(verifica_campo_vazio(str(tabela['Descontos'][contador]), descontos_size) == ""):
    linha = linha + converte_campo_numerico_simples(str(tabela['Descontos'][contador]), descontos_size)
  else:
    linha = linha + verifica_campo_vazio(str(tabela['Descontos'][contador]), descontos_size)

  if(verifica_campo_vazio(str(tabela['Total do Pedido'][contador]), total_pedido_size) == ""):
    linha = linha + converte_campo_numerico_flutuante(str(tabela['Total do Pedido'][contador]), total_pedido_size)
  else:
    linha = linha + verifica_campo_vazio(str(tabela['Total do Pedido'][contador]), total_pedido_size)

  arquivo_saida.write(linha)
  if(contador < tabela['Data Emissao Pedido'].size - 1):
    arquivo_saida.write("\n")
  
  contador = contador + 1
```
